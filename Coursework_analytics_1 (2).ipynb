#### 1.1. Подготовим данные для работы

```python
import pandas as pd
import csv
import os

# константы
# DATA_PATH - путь к папке с данными
DATA_PATH = "Data"

folder_name = os.path.normpath(DATA_PATH)
files_path = {}

# функция поиска всех файлов с данными по пути, с проверкой на разрешение файла
def search_file(folder_name):
        for dirpath, dirnames, filenames in os.walk(folder_name):
            for file in filenames:
                name_and_expansion = os.path.splitext(file)
                full_file_path = os.path.join(dirpath, file)
                if name_and_expansion[1] == '.csv':
                    files_path[name_and_expansion[0]] = [full_file_path, name_and_expansion[1]]

# функция формирования датафрейма из файла по пути
def file_to_dataframe(file_path_content):
    with open(file_path_content, newline='', encoding='utf8') as csvfile:
        reader = csv.DictReader(csvfile)
        content = pd.DataFrame(reader)
    return content

#собираем список с датафреймами файлов
dataframes_from_files = {}
search_file(folder_name)
for file_name, file_path_and_expansion in files_path.items():    
    dataframes_from_files[file_name] = file_to_dataframe(file_path_and_expansion[0]) 
```

```python
# соединяем таблицу 'progresses' с 'progress_phases' по полю 'progress_id' и удаляем дублюрующий столбец 'id'
merge_prog_phases = dataframes_from_files['progresses'] .merge(dataframes_from_files['progress_phases'], \
                                                             left_on='id',right_on='progress_id', how='outer')
merge_prog_phases.drop(['id'], axis='columns', inplace=True)
# в таблице 'courses' у нас незадокументированный столбец без имени, похоже это нумерация, удаляем.
dataframes_from_files['courses'].drop([''], axis='columns', inplace=True)
# соединяем таблицу 'courses' и 'course_contents' по столбцу 'course_id' и так же удаляем дубль 'id'
merge_course_content = dataframes_from_files['courses'].merge(dataframes_from_files['course_contents'], \
                                                             left_on='id',right_on='course_id', how='outer')
merge_course_content.drop(['id'], axis='columns', inplace=True)
# теперь обьеденяем получившиеся датасеты в один, учитываем три столбца 'course_id', 'module_number', 'lesson_number' 
data = pd.merge(merge_prog_phases, merge_course_content, left_on=['course_id', 'module_number', 'lesson_number'], \
                right_on=['course_id', 'module_number', 'lesson_number'], how='outer')

data.head(3)
```

```python
import numpy as np
from dateutil.relativedelta import relativedelta

# найдем кол-во уникальных курсов - это и будет общее кол-во курсов
course_quantity = data['title'].dropna().unique()
# количество модулей на каждом курсе
module_quantity = data[['title', 'module_title']].dropna().drop_duplicates().groupby('title').count()
# сделаем слайс для подсчета кол-ва уроков
data_slice1 = data[['title', 'module_title', 'lesson_number']].dropna().drop_duplicates()
#  количество уроков в каждом модуле на каждом курсе
lesson_quantity = pd.DataFrame(data_slice1.groupby(['title', 'module_title']).count())
lesson_quantity.rename(columns={'lesson_number': 'count'}, inplace=True)
# медианное количество уроков в модуле на каждом курсе
lesson_median = pd.DataFrame(data[['title', 'module_title', \
                                   'lesson_number']].dropna().drop_duplicates()).groupby(['title', 'module_title'])
lesson_median = pd.DataFrame(lesson_median.count())
lesson_median.rename(columns={'lesson_number': 'count'}, inplace=True)
lesson_median = lesson_median.groupby('title').median()

# количество учеников на каждом курсе
students_course_quantity = data[['student_id', 'title']].drop_duplicates().dropna().groupby(['title']).count()
students_course_quantity_field = data[['student_id', 'title', 'field']].drop_duplicates().dropna().groupby(['title', \
                                                                                                        'field']).count()
# подготавливаем данные в таблице 'students'
students_for_age = dataframes_from_files['students'].copy()
students_for_age.drop(['id_', 'city'], axis='columns', inplace=True)
students_for_age.replace('', np.nan, inplace=True)
students_for_age = students_for_age.dropna()
# отсекаем мусорные значения в дате рождения путем ограничения даты рождения
students_for_age = students_for_age[(students_for_age['birthday'] > \
                                     '1970-01-01') & (students_for_age['birthday'] < '2007-01-01')]
# преобразовываем строку в тип даты
students_for_age['birthday'] = students_for_age['birthday'].apply(lambda x: pd.Timestamp(str(x)))
# мы теряем в производительности, но получаем сразу значение возраста из разницы между датами
def f(end):
    r = relativedelta(pd.to_datetime('now'), end) 
    return r.years
students_for_age['age'] = students_for_age['birthday'].apply(f)
# минимальный, максимальный, средний, медианный возраст студентов
students_describe = students_for_age.describe(percentiles=[0.5])
# делаем слайс для расчета студентов по курсам
data_slice2 = data[['student_id', 'course_id', 'title', 'field']].dropna().drop_duplicates()
# мерджим по id студентов и удаляем NANы
students_for_age_on_course = pd.merge(students_for_age, data_slice2, left_on=['id'], \
                right_on=['student_id'], how='inner')
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
students_for_age_on_course = students_for_age_on_course[['title', 'age']].reset_index(drop=True)
students_for_age_on_course = students_for_age_on_course.groupby('title').describe(percentiles=[0.5])
```

```python
# получим выборку данных из общего датафрейма
data_progress_date = data[data['is_homework'] == 'true'][['title', 'progress_id', 'start_date', \
                                                                       'finish_date']].reset_index(drop=True)
# подготовит датафрейм к анализу заменим наны на 2019-07
data_progress_date.fillna('', inplace=True)
data_progress_date = data_progress_date.replace('', pd.to_datetime('2019-07'))
# нас интересует разрез по месяцам преобразуем колонки в таблице
data_progress_date['start_date'] = pd.to_datetime(data_progress_date['start_date']).dt.strftime("%Y-%m")
data_progress_date['finish_date'] = pd.to_datetime(data_progress_date['finish_date']).dt.strftime("%Y-%m")
data_progress_date = data_progress_date.sort_values(['title', 'start_date'])

# так и не придумал как сделать это в пандас, поэтому сделал перебором
import datetime as dt
from dateutil.relativedelta import relativedelta

result_count_dict = {}
# функция проверки ключа словаря, если нет создаем и увеличиваем счетчик на 1
def dict_add(title, date_to_add): 
    try:
        result_count_dict[title][date_to_add]+=1
    except KeyError:
        try:
            result_count_dict[title][date_to_add] = 1
        except KeyError:
            result_count_dict[title] = {date_to_add: 1}     
            
            
# функция которая учитывает, что прогрессия выполнения ДЗ может занимать несколько месяцев
def func1(title, date, stop_d):    
    stop_date = dt.datetime.strptime(stop_d, '%Y-%m').date()
    date = dt.datetime.strptime(date, '%Y-%m').date()           
    while date <= stop_date:
        dict_add(title, date.strftime('%Y-%m'))
        date+=relativedelta(months=+1)
        
        
# основной цикл проверки вариантов начала и окончания ДЗ                
for row in data_progress_date.values:
    if row[3] == row[2]:
        dict_add(row[0], row[3])    
    else:
        func1(row[0], row[2], row[3])
# преобразуем словарь в DataFrame
progress_hw_by_title = pd.DataFrame(result_count_dict).sort_index()
print(progress_hw_by_title)
```

```python
# переименуем колонки для лучшего восприятия
four_ex_data_finish_module.rename(columns={'student_id': 'count_finish'}, inplace=True)
four_ex_data_start_module.rename(columns={'student_id': 'count_start'}, inplace=True)
# смерджим в один датафрейм
four_ex_conversion = four_ex_data_start_module.merge(four_ex_data_finish_module, left_index=True, \
                                                     right_index=True, how='outer')
# подготовим датафрейм для расчета конверсии
four_ex_conversion = four_ex_conversion.reset_index()
four_ex_conversion['module_number'] = four_ex_conversion['module_number'].apply(int)
four_ex_conversion = four_ex_conversion.set_index(['title', 'module_number', 'module_title']).sort_index().reset_index()
curent_course_name = ''
for i in range(four_ex_conversion['module_number'].count()):
    # не все модули с дз, поэтому нужно найти минимальный номер модуля,
    # для правильного расчета конверсии первого модул с дз в уроке
    course_name = four_ex_conversion.at[i, 'title']
    if curent_course_name != course_name:
        curent_course_name = course_name
        min_number_module_in_course = four_ex_conversion[four_ex_conversion['title'] == f'{curent_course_name}']['module_number'].min()    
    # если модуль первый в курсе, то его конверсию я беру равной проценту студентов сдавших финальное дз модуля
    if four_ex_conversion.at[i, 'module_number'] == min_number_module_in_course:
        four_ex_conversion.at[i, 'conversion'] = four_ex_conversion.at[i, 'count_finish'] / four_ex_conversion.at[i, 'count_start'] * 100
    else:
        four_ex_conversion.at[i, 'conversion'] = four_ex_conversion.at[i, 'count_start'] / four_ex_conversion.at[i-1, 'count_finish'] * 100
# конверсия посчитана
four_ex_conversion.head(5)
```
