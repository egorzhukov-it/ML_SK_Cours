1.2. Давайте посмотрим какой канал поиска кадров наиболее успешен и оптимален по цене качеству

```python
sql_tables = """SELECT source_count."Employee Source", source_count.count, costs."Total",
costs."Total" / source_count.count AS "cost/count"
FROM
(SELECT "Employee Source", COUNT("Employee Name") as count
FROM hr_dataset
GROUP BY 1
ORDER BY 2 DESC)
as source_count
JOIN (SELECT "Employment Source", "Total" FROM recruiting_costs) AS costs
ON costs."Employment Source" = source_count."Employee Source"
ORDER BY 4 DESC;"""
coust_for_month = pd.read_sql_query(sql_tables, conn)
coust_for_month
```

1.3. Посмотрим нагрузку на руководителей производства, может есть перегруженные сотрудники

```python
# Посмотрим сколько работает у каждого руководителя людей в подчинение и на каких должностях
sql_tables = """SELECT "Manager Name", "Position", COUNT("Employee Name")
FROM production_staff
WHERE "TermDate" IS NULL
AND
"Manager Name" IS NOT NULL
GROUP BY 1, 2
ORDER BY 1,2,3 DESC;"""
load_on_manager = pd.read_sql_query(sql_tables, conn)
load_on_manager
```

Видим, что Elisa Bramante является руководителем производства. У нее в подчинение находятся Production Manager.
Так же видно, что кол-во подчиненых у всех разное, значит нагрузка тоже разная, так быть не должно.
Давайте найдем сумму людей в подчинение у менеджера и добавим ставку часа.

```python
# Посмотрим сколько работает у каждого руководителя людей в подчинение и на каких должностях
sql_tables = """SELECT People."Manager Name", ManPay."Pay", People.count
FROM
(SELECT "Manager Name", COUNT("Employee Name") AS count,
SUBSTRING("Manager Name", POSITION(' ' in "Manager Name")+1, LENGTH("Manager Name")) AS family
FROM production_staff
WHERE "TermDate" IS NULL
AND
"Manager Name" IS NOT NULL GROUP BY 1)
AS People
JOIN
(SELECT "Pay",
SUBSTRING("Employee Name", 0, POSITION(',' in "Employee Name")) AS family
FROM production_staff
WHERE "TermDate" IS NULL
AND
"Position" = 'Production Manager')
AS ManPay
ON People.family = ManPay.family
ORDER BY 3 DESC;"""
load_on_manager = pd.read_sql_query(sql_tables, conn)
load_on_manager
```

Как видим, меньшее кол-во людей иногда оплачивается больше чем у других, не совсем логично, мб конечно есть продуктовая особенность, команда работает меньшими силами, но на более дорогом проекте. Опять же базовый персонал это не может устраивать, что пораждает конфликты в коллективе, и может являтся причиной текучки.

Вывод: присуствует несправедливая оплата труда сотрудников.
